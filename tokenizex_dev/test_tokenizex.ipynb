```python
from copy import deepcopy

#dev not working, need positional encoding at the first step to take in to an account!

txt = "wikipediawolna"

def text_emb_pos(txt):
    emb_pos = []
    ebm_t = []

    for i in range(1, len(txt)+1):
        et = tokenizer.encode(txt[:i])
        emb_pos.append(list(zip(et, list(range(len(et))))))
        ebm_t.append(et)

    return emb_pos, ebm_t

def create_mask(emb_pos):
    total_len = len(emb_pos)
    res = []
    mask = np.zeros([total_len, total_len])
    for embt, m in zip(emb_pos, mask):
        res.append(embt[-1])
        m[len(res)-1] = 1
        for ie in embt[:-1]:
            for j, je in enumerate(res[:-1]):
                if je == ie:
                    m[j] = 1
                    break

    return res, mask

def create_mask_embedding(txt):
    emb_pos, _ = text_emb_pos(txt)

    res, mask = create_mask(emb_pos)    

    return [e[0] for e in res], mask

emb, mask = create_mask_embedding(txt)


print(emb)
print(mask)
```

```python
from random import random


def untangle_t(emb, mask):
    emb = np.array(emb)
    mask = np.array(mask).astype(bool)
    res = []
    for m in mask:
        res.append(emb[m].tolist())
    return res

def get_imput(txt):
    bi = int(random()*(len(txt)-1000))
    return txt[bi:bi+200]



txt = get_imput(all_txt)
emb, mask = create_mask_embedding(txt)
res = untangle_t(emb, mask)
_, t = text_emb_pos(txt)

print(res)
print(t)
print(mask)
# assert(t == res)

print(len(txt))
```
